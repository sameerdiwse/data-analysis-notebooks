```python
import pandas as pd
from matplotlib import pyplot as plt
shoe_sales_df = pd.read_csv('shoe_sales.csv')
nike_df = shoe_sales_df[shoe_sales_df['brand']=='Nike']
adidas_df = shoe_sales_df[shoe_sales_df['brand']=='Adidas']
nike_df.shape
nike_df.describe()
nike_df.head(5)
```

```python
nike_df[nike_df.sold_qty.isnull()]
```

#### Nike_Analysis

```python
round_val = round(nike_df.sold_qty.median())
```

```python
nike_df.sold_qty.fillna(round_val, inplace=True)
```

```python
nike_df.loc[[32,48]]
```

```python
def plot_qty():
    plt.figure(figsize=(12, 6))

    dates = nike_df['date']

    plt.plot(dates, nike_df['sold_qty'], marker = '+', color='red')
    plt.plot(dates, adidas_df['sold_qty'], marker = '+', color='blue')

    plt.xticks(rotation=45)

plot_qty()
```
