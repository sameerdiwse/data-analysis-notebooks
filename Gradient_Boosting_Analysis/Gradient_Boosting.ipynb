```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ad_spend.csv')
df.head()
```

```python
df.shape
```

```python
data = df.groupby('Season')[['Ad Spend ($)', 'Revenue ($)']].mean()
```

### Plotting bar chart

```python
data.plot(kind='bar');
plt.xticks(rotation=30);
# With this analysis, we can say revenue is more in winter season.
```

### Plotting scatter plot

```python
for season, subdataframe in df.groupby('Season'):
    plt.scatter(subdataframe['Ad Spend ($)'], subdataframe['Revenue ($)'], alpha=0.6)

plt.legend()
```
