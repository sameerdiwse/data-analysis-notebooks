```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('ad_spend.csv')
df.head()
```

```python
df.shape
```

```python
data = df.groupby('Season')[['Ad Spend ($)', 'Revenue ($)']].mean()
```

### Plotting bar chart

```python
data.plot(kind='bar');
plt.xticks(rotation=30);
# With this analysis, we can say revenue is more in winter season.
```

### Plotting scatter plot

```python
for season, subdataframe in df.groupby('Season'):
    plt.scatter(subdataframe['Ad Spend ($)'], subdataframe['Revenue ($)'], alpha=0.6, label=season)

plt.legend();
```

### Encoding the data

```python
# Doing one hot encoding

encoded_df = pd.get_dummies(df, columns=['Season'])
encoded_df.head()
```

### Splitting the data

```python
X = encoded_df.drop(columns=['Revenue ($)'])
y = encoded_df['Revenue ($)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Training Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(X_train, y_train)

linear_model.score(X_test, y_test)
```

### Training Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

DT_model = DecisionTreeRegressor()
DT_model.fit(X_train, y_train)
DT_model.score(X_test, y_test)
```

### Training Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

grad_boost_model = GradientBoostingRegressor(n_estimators=40)
grad_boost_model.fit(X_train, y_train)
grad_boost_model.score(X_test, y_test)
```
